```python
import numpy as np
import pandas as pd
real_dat = pd.read_csv("day2_input.txt", header = None)
real_dat = real_dat.transpose()
dat = pd.DataFrame(data={'col1': [1,9,10,3,2,3,11,0,99,30,40,50]})
real_dat.columns = ['col1']
dat
dat = real_dat
```

```python

# Part 1

pointer = 0
dat.col1[1] = 12
dat.col1[2] = 2


def parse_block (pointer):
    "parse a four-length block"
    if dat.col1[pointer] == 1:
        dat.col1[dat.col1[pointer + 3]] =  dat.col1[dat.col1[pointer + 1]] + dat.col1[dat.col1[pointer + 2]]
    if dat.col1[pointer] == 2:
        dat.col1[dat.col1[pointer + 3]] =  dat.col1[dat.col1[pointer + 1]] * dat.col1[dat.col1[pointer + 2]]


while(dat.col1[pointer] != 99):
    parse_block(pointer)
    pointer = pointer + 4
    
dat
```
